```python
import numpy as np

#inverse final permutation matrix
invrfp = [57,49,41,33,25,17,9,1,
          59,51,43,35,27,19,11,3,
          61,53,45,37,29,21,13,5,
          63,55,47,39,31,23,15,7,
          58,50,42,34,26,18,10,2,
          60,52,44,36,28,20,12,4,
          62,54,46,38,30,22,14,6,
          64,56,48,40,32,24,16,8,          
          ]

#expansion box
ex = [32, 1, 2, 3, 4, 5,
      4, 5, 6, 7, 8, 9,
      8, 9, 10, 11, 12, 13,
      12, 13, 14, 15, 16, 17,
      16, 17, 18, 19, 20, 21,
      20, 21, 22, 23, 24, 25,
      24, 25, 26, 27, 28, 29,
      28, 29, 30, 31, 32, 1]

#inverse permutation matrix
invp = [9, 17, 23, 31,
       13, 28, 2, 18,
       24, 16, 30, 6,
       26, 20, 10, 1,
       8, 14, 25, 3,
       4, 29, 11, 19,
       32, 12, 22, 7,
       5, 27, 15, 21]

#output file contains pairs of blocks of binary output
with open('binciphertexts.txt', 'r') as f:
    output = f.read().splitlines()

expand_out_1_file=open('expand_out_1.txt', 'w') # output of expansion box for 1st block
sbox_in_xor_file=open('sbox_inxor.txt', 'w')
sbox_out_xor_file=open('sbox_outxor.txt', 'w')

for i in range(0,len(output),2):  
    output1 = output[i]
    output2 = output[i+1]

    invfp1 = []
    invfp2 = []
    for j in invrfp:
        invfp1.append(output1[j-1])
        invfp2.append(output2[j-1])
        
    left_half=invfp1[0:32]
    right_half=invfp1[32:]        
    expanded=[]
    for j in ex:
        expanded.append(left_half[j-1])
    sbox_in_xor = []
    sbox_out_xor = []
    
    expansion_out_1 = []
    #inverse final permutation of 1st and second block (?text)
    invfp_xor = np.bitwise_xor(np.array(invfp1).astype(int),np.array(invfp2).astype(int))
    right_half_xor = invfp_xor[32:]
    left_half_xor = invfp_xor[:32]

    for j in invp:
        sbox_out_xor.append(right_half_xor[j-1])
    for j in ex:
        sbox_in_xor.append(left_half_xor[j-1])
        
    sbox_in_xor_file.write(''.join(map(str,sbox_in_xor))+"\n")
    sbox_out_xor_file.write(''.join(map(str,sbox_out_xor))+"\n")
    expand_out_1_file.write(''.join(map(str,expanded))+"\n")
    
expand_out_1_file.close()
sbox_in_xor_file.close()
sbox_out_xor_file.close()
```
